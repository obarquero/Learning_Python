### 1.) Create a sinusoidal signal with frequency 1/5 Hz, and sampling frequency 1/.02 plus some random noise

```python
from scipy import ndimage, misc
```

```python
lena = misc.lena()
type(lena)
```

```python
plt.imshow(lena, cmap = cm.gray)
```

```python
#copy of lena image
noisy_lena = np.copy(lena).astype(np.float)
#represent fft of lena
print(type(noisy_lena))
fft_lena = np.fft.fft2(noisy_lena)
fft_lena_s = np.fft.fftshift(fft_lena)
im = plt.imshow(np.real(fft_lena_s),cmap = cm.gray)
im.set_clim(0,100)
```

```python
#let's add some noise

noisy_lena = noisy_lena + lena.std()*0.8*np.random.standard_normal(lena.shape)
blurred_lena = ndimage.gaussian_filter(noisy_lena,sigma = 3)
median_lena = ndimage.median_filter(blurred_lena, size = 6)

from scipy import signal
wiener_lena = signal.wiener(median_lena,(5,5))

plt.figure(figsize = (15,5))
plt.subplot(141)
plt.imshow(noisy_lena,cmap = cm.gray)
plt.axis('off')
plt.title('noisy')
plt.subplot(142)
plt.imshow(blurred_lena,cmap = cm.gray)
plt.axis('off')
plt.title('gaussian')
plt.subplot(143)
plt.imshow(median_lena,cmap = cm.gray)
plt.axis('off')
plt.title('median')
plt.subplot(144)
plt.imshow(wiener_lena,cmap = cm.gray)
plt.axis('off')
plt.title('wiener')
```

## Regession Line Boston's house price

```python
from sklearn.datasets import load_boston

data = load_boston()
print data.keys()
```

We can see that thera are just over 500 data points:

```python
print data.data.shape
print data.target.shape
```

```python
print data.DESCR
```

```python
plt.hist(data.target)
plt.xlabel('price ($1000)')
plt.ylabel('count')
```

Try some scatter plots 

```python
plt.scatter(data.data[:,5],data.target)
plt.xlabel('room number')
plt.ylabel('price($1000)')

plt.figure()
plt.scatter(data.data[:,10],data.target)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn import preprocessing
from sklearn import cross_validation
#normalization
X = preprocessing.scale(data.data)
print X.shape
X_train, X_test, y_train, y_test = cross_validation.train_test_split(X,data.target, test_size=0.2, random_state=0)
clf = LinearRegression()


clf.fit(X_train, y_train)
predicted = clf.predict(X_test)
plt.scatter(y_test, predicted)
plt.plot([0, 50], [0, 50], '--k')
plt.axis('tight')
plt.xlabel('True price ($1000s)')
plt.ylabel('Predicted price ($1000s)')
print clf.coef_
print("mse is " + str(np.mean((y_test-predicted)**2)))
```

```python
#Let's select variables using Lasso
from sklearn.linear_model import LassoCV

model = LassoCV(cv = 5).fit(X_train, y_train)

print model.coef_
```

```python
pred_lasso = model.predict(X_test)
plt.scatter(y_test,pred_lasso)
plt.plot([0, 50], [0, 50], '--k')
plt.axis('tight')
plt.xlabel('True price ($1000s)')
plt.ylabel('Predicted price ($1000s)')
print("mse is " + str(np.mean((y_test-pred_lasso)**2)))
```
